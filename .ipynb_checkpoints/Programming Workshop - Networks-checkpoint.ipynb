````python
import math
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
````

We're in a *networks* class, remember? :)

Now, we're going to learn how to create, manipulate and visualize networks using the [NetworkX](https://networkx.github.io/) package

First, we need to `import` the library:


````python
import networkx as nx
````

Let's create an empty network, and add some vertices (nodes) and edges (links):

````python
G=nx.Graph()
# create some nodes
G.add_node(0, name='Foo')
G.add_node(1, name='Bar')
G.add_node(2, name='Baz')
G.add_node(3, name='Quux')

# link some of them together
G.add_edge(0, 1)
G.add_edge(1, 2)
G.add_edge(1, 3)
G.add_edge(2, 3)

````

There are more succint ways to create that graph, but that'll do for now. Now let's visualize it. NetworkX integrates with `matplotlib`, which we already learned.

````python
nx.draw(G)
````

## Building a network from a data file

NetworkX provides a method to work with this format, `read_edgelist()`.

````python
KN = nx.read_edgelist('karate_club.txt')

nx.draw(KN)
````

## Reading network files

NetworkX supports [GEXF](http://gexf.net/format/), a standard format for storing graph-like data. We will read the *character co-appearance* network of the movie *Pulp Fiction*, courtesy of our friends from [Movie Galaxies](http://moviegalaxies.com/).


````python
import urllib2 # we need this to get files from the web

graph_data = urllib2.urlopen("http://media.moviegalaxies.com/gexf/660.gexf")
pf_graph = nx.read_gexf(graph_data, relabel=True) # use the GEXF 'label' attribute as node label
````

Our graph is now stored in the `pf_graph` variable. We can use the `nodes()` and `edges()` methods to inspect the data.

````python
print pf_graph.nodes()[:10]
````

````python
print pf_graph.edges()[:10]
````

Who is *Honney Bunny* connected with?

````python
print pf_graph.neighbors('HONEY BUNNY')
````

Let's visualize this network, this time customizing its appearance a bit. We will use an attribute of the nodes provided by the data source, the [modularity class](http://en.wikipedia.org/wiki/Modularity_%28networks%29) to color communities of nodes.

````python
plt.figure(figsize=(15,10))

layout = nx.spring_layout(pf_graph, iterations=300)
colors = ['#1f77b4','#ff7f0e','#2ca02c','#d62728','#9467bd']
nx.draw_networkx_nodes(pf_graph, layout,
                       node_color=[colors[pf_graph.node[n]['Modularity Class'] % len(colors)] 
                                   for n in pf_graph.nodes()])
nx.draw_networkx_labels(pf_graph, layout, { k: k for k, _ in pf_graph.node.items() }, font_size=14)
nx.draw_networkx_edges(pf_graph, layout)
````

## Analyzing networks

Sometimes it is useful to generate random networks. NetworkX provides [functions to generate them under different models](http://networkx.lanl.gov/reference/generators.html), such as Erdos-Renyi, Watts-Strogratz and Barabasi-Albert.

Let's generate a graph that we're already familiar with, the [Barabasi-Albert](http://en.wikipedia.org/wiki/Barab%C3%A1si%E2%80%93Albert_model) network. Recall that the generation model has two parameters: $m_0$ (number of initial nodes), and $m$ (number of edges to attach from a new node to existing nodes).

````python
ba = nx.barabasi_albert_graph(10000, 5, seed = None)
````

This model generates *scale free* networks, which are networks that exhibit a degree distribution following a power law. We've just learned to how plot data *and* how to create Barabasi-Albert networks, so now let's witness a power-law distribution in all its glory.

````python
degs = nx.degree(ba).values()
degs = np.array(degs)
print degs
````

````python
kmin=min(degs)
kmax=max(degs)
nbins=10
dk=(kmax-kmin)/(nbins-1)
C=[kmin+i*dk for i in range(0,nbins)]
print C
````

````python
N=[0]*(len(C)-1)
for i,c in enumerate(C[:-1]):
    N[i]=len(degs[(degs>=c)&(degs<C[i+1])])
N[-1]+=len(degs[degs==kmax])
N=[float(n)/sum(N) for n in N]
````

````python
plt.loglog(C[:-1],N)
````

````python
logdegs=np.log(degs)
````

````python
logkmin=min(logdegs)
logkmax=max(logdegs)
nbins=10
dk=(logkmax-logkmin)/(nbins-1)
C=[logkmin+i*dk for i in range(0,nbins)]
N=[0]*(len(C)-1)
for i,c in enumerate(C[:-1]):
    N[i]=len(logdegs[(logdegs>=c)&(logdegs<C[i+1])])
N[-1]+=len(logdegs[logdegs==logkmax])
````

````python
N=N/np.diff(np.exp(C))
N=N/sum(N)
````

````python
plt.loglog(C[:-1],N)
````
